```python
#Making 'early morning','morning','noon','evening' etc. from 'dep_time_hour' feature for better prediction.
def flight_dep_time(x):
    if (x>4) and (x<=8):
        return "Early Morning"
        
    elif (x>8) and (x<=12):
        return "Morning"
        
    elif (x>12) and (x<=16):
        return "Noon"
        
    elif (x>16) and (x<=20):
        return "Evening"
        
    elif (x>20) and (x<=24):
        return "Night"
    else:
        return "late Night"
```

```python
#I have to do pre processing - converting duration feature into separate fromet like 'h' and 'm' hours and minutes.like convert 50m into 0h and 50m.
# Data cleaning and preparation
#convert Duration feature into separate like 2 hour 30 min, 00 hour 45 min.

def preprocess_duration (x):
    if 'h' not in x:
        x = '0h' + ' ' + x
    elif 'm' not in x:
        x = x + ' ' + '0m'
    return x
```

```python
train_data['Duration'] = train_data['Duration'].apply(preprocess_duration)
```

```python
#Removing outliers 
#First step is outlier detaction.
#Outliers are those vaues which are extremely low and extremely high
#for detecting outliers the best technique is visualization with diffrent plots.

def plot(df, col):
    fig, (ax1, ax2, ax3) = plt.subplots(3,1) #3 rows and 1 column

    sns.distplot(df[col], ax=ax1)
    sns.boxplot(df[col], ax=ax2)
    sns.histplot(df[col], ax=ax3, kde=False)
```

```python
#1st step is Prepare data for making ARIMA Model for feature 'Journey_Day' and target variable is 'Price'.

import pandas as pd

# Group by Journey_day and calculate the mean price for each day
daily_price = train_data.groupby('Journey_day')['Price'].mean().reset_index()

# Set Journey_day as the index (numeric, sequential)
daily_price.set_index('Journey_day', inplace=True)
```

```python
#2nd step is Stationarity Check for 'Price' column

from statsmodels.tsa.stattools import adfuller

result = adfuller(daily_price['Price'])
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')

if result[1] > 0.05:
    print("The data is not stationary. Differencing is required.")
else:
    print("The data is stationary.")
```

```python
#3rd Fit ARIMA Model to predict the Price

from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Define ARIMA order (p, d, q) -  I can tune these values as needed
order = (1, 1, 1)

# Fit the ARIMA model
model = ARIMA(daily_price['Price'], order=order)
results = model.fit()

# Print summary
print(results.summary())

# Plot actual vs fitted values
plt.figure(figsize=(10, 5))
plt.plot(daily_price.index, daily_price['Price'], label='Actual Prices')
plt.plot(daily_price.index, results.fittedvalues, color='red', label='Fitted Prices')
plt.title('ARIMA Model - Journey_day vs Price')
plt.xlabel('Journey_day')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
#Forecast Future Prices - Predict prices for future days

# Forecast the next 5 days
forecast = results.forecast(steps=5)
print("Forecasted Prices for next 5 days:", forecast)

# Plot forecast
future_days = range(daily_price.index[-1] + 1, daily_price.index[-1] + 6)  # Next 5 days
plt.figure(figsize=(10, 5))
plt.plot(daily_price.index, daily_price['Price'], label='Historical Prices')
plt.plot(future_days, forecast, color='green', label='Forecasted Prices')
plt.title('ARIMA Model - Forecast')
plt.xlabel('Journey_day')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
#Evaluate Model Performance - Calculate performance metrics for the model:

from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Calculate metrics
mae = mean_absolute_error(daily_price['Price'], results.fittedvalues)
rmse = np.sqrt(mean_squared_error(daily_price['Price'], results.fittedvalues))

print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
```
